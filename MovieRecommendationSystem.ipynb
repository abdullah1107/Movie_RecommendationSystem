```python
import pandas as pd
```

```python
rating_movies = pd.read_csv("ml-latest-small/ratings.csv")
```

```python
rating_movies.head(10)
```

```python
movies_name= pd.read_csv("ml-latest-small/movies.csv")
```

```python
movies_name.head(10)
```

```python
movies_data = pd.merge(rating_movies, movies_name, on = 'movieId')
```

```python
movies_data.head(10)
```

```python
movies_data.groupby('title')['rating'].mean().sort_values(ascending = False).head(10)
```

```python
movies_data.groupby('title')['rating'].count().sort_values(ascending = False).head(10)
```

```python
ratings = pd.DataFrame(movies_data.groupby('title')['rating'].mean().sort_values(ascending = False))
ratings.head(10)
```

```python
ratings['number_of_ratings'] = pd.DataFrame(movies_data.groupby('title')['rating'].count())
ratings.head(10)
```

```python
ratings.sort_values('number_of_ratings',ascending=False).head(10)
```

```python
user_movie_rating = movies_data.pivot_table(index='userId', columns='title', values='rating')
user_movie_rating.head()
```

```python
ratings.sort_values('number_of_ratings',ascending=False).head(10)
```

```python
matrix_movie_rating = user_movie_rating['Matrix, The (1999)']
matrix_movie_rating.head(10)
```

```python
similler_matrix_movie_rating = user_movie_rating.corrwith(matrix_movie_rating)
```

```python
corr_matrix_movie = pd.DataFrame(similler_matrix_movie_rating, columns = ['Correlation'])
corr_matrix_movie.dropna(inplace  = True)
corr_matrix_movie.head(10)
```

```python
corr_matrix_movie.sort_values('Correlation', ascending = False).head(10)
```

```python
corr_matrix_movie = corr_matrix_movie.join(ratings['number_of_ratings'])
corr_matrix_movie.head(10)
```

```python
corr_matrix_movie[corr_matrix_movie['number_of_ratings']>100].sort_values('Correlation',ascending=False).head()
```
